```python
# Visualize the count of ContractRenewal
plt.figure(figsize=(8, 6))
sns.countplot(x='ContractRenewal', data=data, palette='viridis')
plt.title('Count of Contract Renewal', fontsize=15)
plt.xlabel('Contract Renewal', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.xticks(rotation=0)
plt.show()
```

```python
sns.countplot(x='DataUsage', data=data, order=data['DataUsage'].value_counts().index, color='cornflowerblue')
plt.xticks(rotation=90)
```

Pie Chart

Univariate Data Visualisation - Numerical Data

```python
# Pie chart for Churn
churn_counts = data['Churn'].value_counts()

plt.figure(figsize=(8, 8))
plt.pie(churn_counts, labels=churn_counts.index, autopct='%1.1f%%', startangle=90, colors=['lightgreen', 'salmon'])
plt.title('Proportion of Customers by Churn', fontsize=15)
plt.show()
```

```python
# Pie chart for ContractRenewal
renewal_counts = data['ContractRenewal'].value_counts()

plt.figure(figsize=(8, 8))
plt.pie(renewal_counts, labels=renewal_counts.index, autopct='%1.1f%%', startangle=90, colors=['skyblue', 'lightcoral'])
plt.title('Proportion of Customers by Contract Renewal', fontsize=15)
plt.show()
```

```python
# Pie chart for DataPlan
dataplan_counts = data['DataPlan'].value_counts()

plt.figure(figsize=(8, 8))
plt.pie(dataplan_counts, labels=dataplan_counts.index, autopct='%1.1f%%', startangle=90, colors=['gold', 'lightgray'])
plt.title('Proportion of Customers by Data Plan', fontsize=15)
plt.show()
```
